# Subjective Evaluations

```python
import pandas as pd
from scipy.stats import f_oneway
```

```python
resp_data = pd.read_csv('questionnaire\\response_summary.csv')
resp_data.head(5)
```

```python
columns_to_compare = ['competence', 'activity', 'frustration', 'pleasure', 'pressure', 'effort', 'immersion', 'physicality', 'difficulty', 'familarity', 'understanding']

for column in columns_to_compare:
    grouped_data = resp_data.groupby('method')[column].agg(['mean', 'std'])
    anova_result = f_oneway(*[group[column] for name, group in resp_data.groupby('method')])
    print(f"Mean and Standard Deviation of '{column}' by 'method':")
    print("\nANOVA Test Result:")
    print("F-value:", anova_result.statistic)
    print("p-value:", anova_result.pvalue)
    print("\n")
```
