Verify whether the distribution of group data follows normal distribution.

```python
import pandas as pd
from scipy.stats import shapiro

# Load the data from the CSV file
data_path = 'questionnaire\\short_term\\response_summary_short.csv'
df = pd.read_csv(data_path)

# Group the data by 'strategy' and 'method'
grouped = df.groupby(['strategy', 'method'])

# Perform the Shapiro-Wilk test for normality on each group and each numeric column
results = {}
for (strategy, method), group in grouped:
    for column in group.select_dtypes(include=['float64', 'int64']).columns:
        if column in ['strategy','topic','corpus','method','participant']:
            continue
        if column not in results:
            results[column] = {}
        stat, p_value = shapiro(group[column])
        results[column][(strategy, method)] = p_value

# Display the results
for column, groups in results.items():
    print(f"Column: {column}")
    for group, p_value in groups.items():
        print(f"  Group: {group} - p-value = {p_value:.4f} {'(Not Normal)' if p_value < 0.05 else '(Normal)'}")
    print()
```
